# What are the ten most common movie names of all time

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV files into Pandas DataFrames
cast_df = pd.read_csv('cast.csv')
release_dates_df = pd.read_csv('release_dates.csv')
titles_df = pd.read_csv('titles.csv')

# Task 1: Find the ten most common movie names of all time
top_10_common_titles = titles_df['title'].value_counts().head(10)
print("Ten most common movie names of all time:")
print(top_10_common_titles)


```

```python
# Task 2: Find three years of the 1930s with the most films released
films_1930s = release_dates_df[(release_dates_df['year'] >= 1930) & (release_dates_df['year'] <= 1939)]
most_films_1930s = films_1930s['year'].value_counts().head(3)
print("\nThree years of the 1930s with the most films released:")
print(most_films_1930s)


```

```python
# Task 3: Plot the number of films released each decade over the history of cinema
release_dates_df['decade'] = release_dates_df['year'] // 10 * 10
films_per_decade = release_dates_df['decade'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
films_per_decade.plot(kind='bar', xlabel='Decade', ylabel='Number of Films', title='Number of Films Released Each Decade')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
How many leading roles(n=1) were available to actors and how many to actresses, in the 1950s?
```

```python
leading_roles_1950s = cast_df[(cast_df['year'] >= 1950) & (cast_df['year'] <= 1959) & (cast_df['n'] == 1)]
leading_roles_actors_1950s = leading_roles_1950s[leading_roles_1950s['type'] == 'actor']['name'].count()
leading_roles_actresses_1950s = leading_roles_1950s[leading_roles_1950s['type'] == 'actress']['name'].count()
print("\nLeading roles (n=1) in the 1950s:")
print("Actors:", leading_roles_actors_1950s)
print("Actresses:", leading_roles_actresses_1950s)
```

```python
How many supporting roles(n=2) were available to actors and how many to actresses, in the 1950s?
```

```python
roles_1950s = cast_df[(cast_df['year'] >= 1950) & (cast_df['year'] <= 1959)]
Supporting_roles_actors = roles_1950s[(roles_1950s['n'] == 2) & (roles_1950s['type'] == 'actor')].shape[0]
Supporting_roles_actresses = roles_1950s[(roles_1950s['n'] == 2) & (roles_1950s['type'] == 'actress')].shape[0]
print("\nsupporting roles (n=2) in the 1950s:")
print("Actors:", Supporting_roles_actors)
print("Actresses:", Supporting_roles_actresses)
```
