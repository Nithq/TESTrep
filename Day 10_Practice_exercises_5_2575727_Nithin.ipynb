```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV files
cast_df = pd.read_csv('cast.csv')
release_dates_df = pd.read_csv('release_dates.csv')
titles_df = pd.read_csv('titles.csv')
```

# 1.Make a bar plot of the months in which movies with "Christmas" in their title tend to be released in the USA

```python
# 1. Make a bar plot of the months in which movies with "Christmas" in their title are released in the USA
christmas_movies_usa = release_dates_df[(release_dates_df['country'] == 'USA') & (release_dates_df['title'].str.contains('Christmas'))]
christmas_movies_usa['date'] = pd.to_datetime(christmas_movies_usa['date'])
christmas_movies_usa['month'] = christmas_movies_usa['date'].dt.month
month_counts = christmas_movies_usa['month'].value_counts().sort_index()
month_counts.plot(kind='bar', title='Christmas Movies Release Months in the USA')
plt.xlabel('Month')
plt.ylabel('Number of Movies')
plt.show()
```

# 2.Make a bar plot of the months in which movies whose titles start with "The Hobbit" are released in the USA

```python
# 2. Make a bar plot of the months in which movies whose titles start with "The Hobbit" are released in the USA
the_hobbit_movies_usa = release_dates_df[(release_dates_df['country'] == 'USA') & (release_dates_df['title'].str.startswith('The Hobbit'))]
the_hobbit_movies_usa['date'] = pd.to_datetime(the_hobbit_movies_usa['date'])
the_hobbit_movies_usa['month'] = the_hobbit_movies_usa['date'].dt.month
month_counts = the_hobbit_movies_usa['month'].value_counts().sort_index()
month_counts.plot(kind='bar', title='The Hobbit Movies Release Months in the USA')
plt.xlabel('Month')
plt.ylabel('Number of Movies')
plt.show()
```

# 3.Make a bar plot of the day of the week on which movies with "Romance" in their title tend to be released in the USA

```python
# 3. Make a bar plot of the day of the week on which movies with "Romance" in their title are released in the USA
romance_movies_usa = release_dates_df[(release_dates_df['country'] == 'USA') & (release_dates_df['title'].str.contains('Romance'))]
romance_movies_usa['date'] = pd.to_datetime(romance_movies_usa['date'])
romance_movies_usa['day_of_week'] = romance_movies_usa['date'].dt.dayofweek
day_counts = romance_movies_usa['day_of_week'].value_counts().sort_index()
day_names = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
day_counts.index = day_names
day_counts.plot(kind='bar', title='Romance Movies Release Days in the USA')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Movies')
plt.show()
```
