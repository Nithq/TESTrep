```python
import pandas as pd

# Load the CSV files for cast, release dates, and titles
cast = pd.read_csv('cast.csv')
release_dates = pd.read_csv('release_dates.csv')
titles = pd.read_csv('titles.csv')
```

# Using groupby(), plot the number of films that have been released each decade in the history of cinema

```python
#Task 1: Plot the number of films released each decade
release_dates['decade'] = release_dates['year'] // 10 * 10
film_counts = release_dates['decade'].value_counts().sort_index()
film_counts.plot(kind='bar', title='Number of Films Released Each Decade')
```

# Use groupby() to plot the number of "Hamlet" films made each decade

```python
release_dates['decade'] = release_dates['year'] // 10 * 10
film_counts = release_dates['decade'].value_counts().sort_index()
film_counts.plot(kind='bar', title='Number of Hamlet Films Released Each Decade')
```

# How many leading roles(n=1) were available to actors and how many to actresses, in each year of the 1950s?

```python
# Task 3: Count leading roles for actors and actresses in the 1950s
roles_1950s = cast[(cast['year'] // 10 == 195) & (cast['n'] == 1)]
roles_by_gender = roles_1950s['type'].value_counts()
print("Leading roles for actors and actresses in the 1950s:")
print(roles_by_gender)
```

# In the 1950s decade taken as a whole, how many total roles were available to actors and how many to actresses for each "n" number 1 through 5

```python
# Task 4: Count total roles for actors and actresses by "n" in the 1950s
roles_by_n = cast[(cast['year'] // 10 == 195) & (cast['n'] <= 5)].groupby(['n', 'type']).size().unstack().fillna(0)
print("Total roles for actors and actresses in the 1950s by 'n':")
print(roles_by_n)
```

# Use groupby() to determine how many roles are listed for each of the pink panther movies

```python
# Task 5: Count roles in each Pink Panther movie
pink_panther_roles = cast[cast['title'].str.contains('Pink Panther')].groupby('title').size()
print("Roles in each Pink Panther movie:")
print(pink_panther_roles)
```

# List in order by year each of the films in which Frank Oz has played more than 1 role

```python
# Task 6: Films where Frank Oz played more than 1 role
frank_oz_films = cast[cast['name'] == 'Frank Oz'].groupby(['title', 'year']).size()
frank_oz_multi_role_films = frank_oz_films[frank_oz_films > 1]
print("Films in which Frank Oz played more than 1 role:")
print(frank_oz_multi_role_films)

```

# List each of the characters that Frank Oz has portrayed at least twice

```python
# Task 7: Characters portrayed by Frank Oz at least twice
frank_oz_characters = cast[cast['name'] == 'Frank Oz'].groupby('character').size()
frank_oz_multi_characters = frank_oz_characters[frank_oz_characters > 1]
print("Characters portrayed by Frank Oz at least twice:")
print(frank_oz_multi_characters)
```
